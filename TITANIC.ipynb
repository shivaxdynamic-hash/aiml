```python
#importing necessary libraries
import pandas as pd
import numpy as np

from scipy import stats
from sklearn.preprocessing import LabelEncoder
```

```python
#loading dataset
df=pd.read_csv("/content/drive/MyDrive/titanic.csv")

#first 5 rows
print(df.head())

#about dataset
print(df.info())

#statistical summary
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
print(df['Age'].isnull().sum())
print(df['Embarked'].isnull().sum())
#now we can use dropna
```

```python
#using drop and dropna
df.drop(['Cabin'],axis=1,inplace=True)
df.dropna(subset=['Embarked'],inplace=True)
print(df.isnull().sum())
print(df.info())
```

```python
#using fiilna for age column
fillna_cols=['Age']
for col in fillna_cols:
 df[col]=df[col].fillna(df[col].mean())
print(df.isnull().sum())
```

```python
#duplicates
print(df.duplicated().sum())
print(df.shape)
```

```python
#labelencoding
le=LabelEncoder()
categorical_cols=df.select_dtypes(include=['object']).columns
for col in categorical_cols:
  df[col]=le.fit_transform(df[col])
print(df.head())

```

```python
#separate input and output
X=df.drop(['Survived'],axis=1)
Y=df['Survived']
print("input columns:",X.head())
print("output column:",Y.head())
print(df.loc[2,'Name'])
```

```python
#train and test
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print("X-train shape: ",X_train.shape)
print("X-test shape: ",X_test.shape)
print("Y-train shape: ",Y_train.shape)
print("Y-test shape: ",Y_test.shape)
```

```python
#featurescaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
print(X_train)
```

```python
#model selction
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC()
}

for name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    acc = (Y_pred == Y_test).mean()#find fraction of all true values
    print(f"{name}: {acc:.4f}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,precision_score,recall_score,f1_score

cm = confusion_matrix(Y_test, Y_pred)
print(cm)
print(classification_report(Y_test, Y_pred))  # precision, recall, f1, support
print(accuracy_score(Y_test, Y_pred))  # accuracy
```
